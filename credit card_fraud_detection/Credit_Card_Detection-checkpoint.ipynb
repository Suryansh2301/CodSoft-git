```python
def haversine_distance(lat1, lon1, lat2, lon2):
    r = 6371 # earth radius in km

    lat1, lon1, lat2, lon2 = map(np.radians,[lat1, lon1, lat2, lon2])

    dlat =lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2*np.arcsin(np.sqrt(a))
    return r*c
```

1) Classification Metrics:

This is a major focus, offering metrics like:

2) Accuracy Score: Measures the proportion of correctly classified instances.
3) Precision, Recall, F1-score: These metrics are particularly useful for imbalanced datasets, offering insights into the model's ability to correctly identify positive instances (precision) and capture all positive instances (recall), and a harmonic mean of both (F1-score).
4) Confusion Matrix: A table summarizing the performance of a classification model, showing true positives, true negatives, false positives, and false negatives.
5) ROC AUC Score: Computes the Area Under the Receiver Operating Characteristic (ROC) curve, useful for evaluating binary classifiers across various thresholds.
6) Classification Report: Generates a text report showing the main classification metrics for each class
